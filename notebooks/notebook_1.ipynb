```python
def neigh_homogeneity(df,radius = 2000):
    nh = []
    for i in range(len(df)):
        nx = abs(df['X'].values-df['X'].iloc[i])<=radius
        ny = abs(df['Y'].values-df['Y'].iloc[i])<=radius
        n = df[np.logical_and(nx,ny)]['km1'] # vettore di tutte le cluster labels del neighbourhood
        homogeneity = sum(n==df['km1'].iloc[i])/len(n) # proporzione di labels del neighbourhood che sono uguali a quelle del punto X,Y
        nh.append(homogeneity) 
    
    return nh
```

```python
df_amsterdam['nh_homogeneity']= neigh_homogeneity(df_amsterdam)
```

```python
sns.histplot(df_amsterdam['nh_homogeneity'])
```
